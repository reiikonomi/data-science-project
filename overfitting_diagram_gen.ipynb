```python
import matplotlib.pyplot as plt

# Example data (replace with actual training/validation metrics)
epochs = range(1, 11)  # Assuming 10 epochs
train_acc = [0.75, 0.82, 0.88, 0.92, 0.95, 0.97, 0.98, 0.99, 0.99, 0.99]  # Training accuracy
val_acc = [0.70, 0.78, 0.82, 0.85, 0.87, 0.88, 0.89, 0.88, 0.87, 0.86]  # Validation accuracy
train_loss = [0.5, 0.4, 0.3, 0.25, 0.2, 0.15, 0.12, 0.1, 0.08, 0.07]  # Training loss
val_loss = [0.7, 0.6, 0.5, 0.45, 0.4, 0.38, 0.36, 0.35, 0.34, 0.33]  # Validation loss
```

```python
plt.figure(figsize=(12, 6))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(epochs, train_acc, 'bo-', label='Training Accuracy')
plt.plot(epochs, val_acc, 'ro-', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(epochs, train_loss, 'bo-', label='Training Loss')
plt.plot(epochs, val_loss, 'ro-', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
